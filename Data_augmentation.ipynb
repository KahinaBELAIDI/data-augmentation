```python

import numpy as np
import tensorflow as tf

from matplotlib import pyplot as plt
```

```python
image_path = 'dog.jpg'

plt.imshow(plt.imread(image_path))
```

```python
img = plt.imread(image_path)
```

# Rotation

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=40
)
```

```python
x = next(generator.flow(np.expand_dims(img, axis=0), batch_size=1))
plt.imshow(x[0,:].astype('uint8'))
```

# Width and Height Shifts

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    
    height_shift_range=-0.2
)
```

```python
y = next(generator.flow(np.expand_dims(img, axis=0), batch_size=1))
plt.imshow(y[0].astype('uint8'))
```

# Brighteness

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    brightness_range=(0., 2.)
)

x = next(generator.flow(np.expand_dims(img, axis=0), batch_size=1))
plt.imshow(x[0].astype('uint8'))
```

# Zoom

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    zoom_range=0.5

)

x= next(generator.flow(np.expand_dims(img, axis=0), batch_size=1))
plt.imshow(x[0,:].astype('uint8'))
```

# Flip

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    horizontal_flip=True,
    vertical_flip=True
)

x = next(generator.flow(np.expand_dims(img, axis=0), batch_size=1))
plt.imshow(x[0].astype('uint8'))
```

# Rescale

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255
)
x = next(generator.flow(np.expand_dims(img, axis=0), batch_size=1))
plt.imshow(x[0])
```
